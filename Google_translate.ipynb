```python
import pandas as pd 
import numpy as np 
import seaborn as sns

# convert words into array form or in vector form
from sklearn.feature_extraction.text import CountVectorizer  

# to spilt datasets for train and test 
from sklearn.model_selection import train_test_split 

# to count word 
from sklearn.naive_bayes import MultinomialNB
```

```python
# uploading datasets as df
df = pd.read_csv('language.csv')
```

```python
df
```

```python
# cleaning and filling null values 
df.isna().sum()
```

<h3> preparing data for model

```python
x = np.array(df['Text'])
y = np.array(df['language'])
```

```python
print(x)
```

```python
print(y)
```

```python
# convert words into vector form to train model 
cv = CountVectorizer()
X =cv.fit_transform(x)
```

```python
#spiliting datasets into train and test for model

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#checking shape of train and test
X_train.shape , X_test.shape
```

```python
y_test
```

```python
# calling alorithm naive bayes 
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
# this 95% this model learn from giving data

model.score(X_test,y_test)
```

```python
#create a function to take input and gives output as which language we give as input 
user = input('Enter a Test')
df = cv.transform([user]).toarray()
output = model.predict(df)
print(output)
```
